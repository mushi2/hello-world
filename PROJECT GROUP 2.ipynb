THE GRAPH SHOWING NUMBER OF DEATHS CAUSED BY MALARIA AND PNEUMONIA IN DIFFERENT REGIONS

```python
plt.plot(data2.Malaria, data2.Region)
plt.plot(data2.Pneumonia, data2.Region)
plt.legend(['Malaria', 'Pneumonia'])
plt.title("THE GRAPH SHOWING NUMBER OF DEATHS CAUSED BY MALARIA AND PNEUMONIA")
plt.ylabel('REGIONS')
plt.xlabel('NUMBER OF DEATHS')
plt.grid()
plt.show()

```
